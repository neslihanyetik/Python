```python
Given a string of words, you need to find the highest scoring word.
Each letter of a word scores points according to its position in the alphabet: a = 1, b = 2, c = 3 etc.
You need to return the highest scoring word as a string.
If two words score the same, return the word that appears earliest in the original string.
All letters will be lowercase and all inputs will be valid.
Examples:
'man i need a taxi up to ubud' ---> 'taxi'
'what time are we climbing up the volcano' ----> 'volcano'
'take me to semynak' ---> 'semynak'
'aa b'  --->  'aa'
'b aa'  --->  'b'
'bb d'  --- > 'bb'
'd bb'  ---> 'd'
"aaa b"  -->  "aaa"
```

```python
alphabet = [chr(i) for i in range(97, 123)]
dictt_1 = {}
dictt_2 = {}
for i in range(len(alphabet)):
    dictt_1[alphabet[i]] = i + 1
text = input('Please enter your text here > ')
text_list = text.lower().split()
for i in text_list:
    total = 0
    for x in i:
        total += dictt_1[x]
    dictt_2[i] = total
listt = list(dictt_2.values())
biggest = max(listt)
for x, y in dictt_2.items():
    if y == biggest:
        print(x)
        break
```

```python
sentence = [input("Please enter a sentence").lower()]
sen_list = sentence[0].split()
letters_point = {}
letters = ["a","b","c","d","e","f","g","h","Ä±","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
i = 0
k = 1
while i < 26:
  a = letters[i]
  letters_point[a] = k
  i += 1
  k += 1

total_word ={} 
for word in sen_list:
  total_point = 0
  for item in word:
    total_point += letters_point[item]
  total_word[word] = total_point

max_word = max(total_word, key = total_word.get)
print(max_word)
```

```python
def chrr(n):
    return [i for i in n]

x = list(enumerate([chr(i) for i in range(97,123)] ,1))
y = "what time are we climbing up the volcano".split()
z = list(map(chrr ,y))

for k in z:
    for i in range(len(k)):
        for j,t in x:
            if t == k[i]:
                k[i]=j

m =[sum(i) for i in z]

u = m.index(max(m))
print(y[u])
```

```python
import string
alphabet_string = string.ascii_lowercase
alphabet_list = list(alphabet_string)
alphebet_point=list(enumerate(alphabet_list,1))
alphebet_dict={}
for x,y in alphebet_point:
  alphebet_dict[y]=x
phrase=input('Enter a phrase to be evaluated: ')
phrase_list=phrase.split()
phrase_dict={}
for i in phrase_list:
  point_i=0
  for ii in list(i):
    point_i=point_i+alphebet_dict[ii]
  phrase_dict[i]=point_i
max_word=max(phrase_dict, key=phrase_dict.get)
print(f"'{max_word}' has the highest score")
```

```python
sentence = 'what time are we climbing up the volcano'
list_sen = sentence.split()
alphabet = [chr(i) for i in range(97, 123)]
enumerate(alphabet)
dict_num = dict(enumerate(alphabet, start=1))
alph_num_dict = {}
for a, b in dict_num.items() :
  alph_num_dict[b] = a
print(alph_num_dict)
count_dict = {}
for word in list_sen:
    y = 0
    for i in word:
        x = alph_num_dict[i]
        count_dict[word] = x + y
        y = x
print(count_dict)
for a, b in count_dict.items() :
    if b == max(count_dict.values()) :
      print(f"{a} has the max score and its score is {b}")
```
