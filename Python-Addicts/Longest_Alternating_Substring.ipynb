## Longest Alternating Substring
Given a string of digits, return the longest substring with alternating odd/even or even/odd digits. If two or more substrings have the same length, return the substring that occurs first.

### https://edabit.com/challenge/RB6iWFrCd6rXWH3vi

```python
def substring(string):
  list_substring = []
  k = 0
  substring =""
  while k < len(string) - 1:
    if k == len(string) -2 :
      if int(string[k])%2 != int(string[k+1])%2:
        substring += string[k:]
        list_substring.append(substring)
    elif int(string[k])%2 != int(string[k+1])%2:
      substring += string[k]
    else:
      substring += string[k]
      if len(substring) > 1:
        list_substring.append(substring)
      substring = ""
    k += 1
  print(list_substring)

  max_lst =[]
  for i in list_substring:
    max_lst.append(len(i))
  for k in list_substring:
    if len(k) == max(max_lst):
      return k

substring("721449827599186159274227324466")
```

```python
def alternating(text):
    liste = []
    if int(text[0]) % 2 == 0:
        odd = True
    else:
        odd = False
    if int(text[-1]) % 2 == 0:
        text += '0'
    else:
        text += '1'
    yeni = text[0]
    for i in text[1:]:
        if odd:
            if int(i) % 2 != 0:
                yeni += i
                odd = False
            else:
                if len(yeni) > 1:
                    liste.append(yeni)
                yeni = str(i)
        else:
            if int(i) % 2 == 0:
                yeni += i
                odd = True
            else:
                if len(yeni) > 1:
                    liste.append(yeni)
                yeni = str(i)
    if liste != []:
        result = liste[0]
        for i in liste:
            if len(i) > len(result):
                result = i
    else:
        result = liste
    return liste, f'The longest alternating substring is: {result}'

alternating("594127169973391692147228678476")
```
